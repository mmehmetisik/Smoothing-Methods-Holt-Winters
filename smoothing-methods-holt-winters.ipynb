<h3 style="text-align: center; color: #2c3e50; font-size: 1.8em; border-bottom: 3px double #2980b9; padding-bottom: 5px; margin-bottom: 20px;">üîç What is the Smoothing Method?</h3>

<p style="font-size: 1.2em; color: #333; margin-top: 20px;">The smoothing method is a technique frequently used for time series analysis and forecasting. The primary purpose of this method is to reduce the noise (random fluctuations) in the data, highlighting the underlying main trend or pattern. This is essential for making more accurate and meaningful predictions.</p>

<h4 style="color: #27ae60; font-size: 1.6em; margin-top: 30px; border-left: 5px solid #27ae60; padding-left: 15px;">üìö Smoothing Methods Categories</h4>

<ul style="list-style-type: none; padding: 0; margin-top: 15px; font-size: 1.1em;">
    <li style="padding: 8px 0; color: #34495e;"><strong>Moving Average Method:</strong> A specific window width is chosen, and the average of the data points within this window is taken to form a new data point. This window is then slid across the dataset to compute moving averages for the entire set.</li>
    <li style="padding: 8px 0; color: #34495e;"><strong>Exponential Smoothing Method:</strong> New predictions are formulated by taking the weighted average of past data points. In this method, more recent data is given higher weight, while older data receives lower weight. This is based on the assumption that the data is more related to its recent past.</li>
    <li style="padding: 8px 0; color: #34495e;"><strong>Further Divisions in Exponential Smoothing:</strong> The exponential smoothing method is further divided into Single, Double, and Triple Exponential Smoothing. These methods are used to model different components (level, trend, and seasonality) in the dataset.</li>
</ul>

<p style="font-size: 1.2em; color: #333; margin-top: 20px; border-top: 2px dashed #bdc3c7; padding-top: 20px;">üåê Smoothing methods are commonly used in various domains requiring time series forecasting, such as stock markets, energy consumption, and weather forecasting.</p>

<h3 style="color: #3498db; font-size: 1.8em; margin-top: 30px;">Single Exponential Smoothing (SES)</h3>
<ul style="font-size: 1.2em; color: #333; margin-top: 20px;">
    <li>This method is suitable only for stationary series. It should neither have a trend nor seasonality.</li>
    <li>It can only model the 'Level' component (a concept similar to the average of the series).</li>
    <li>The prediction of the future is more related to the data of the recent past. Under this assumption, predictions are formed by weighting past data.</li>
    <li>This method is quite successful for series that do not contain trends or seasonality.</li>
</ul>

<h3 style="color: #3498db; font-size: 1.8em; margin-top: 30px;">Double Exponential Smoothing (DES)</h3>
<ul style="font-size: 1.2em; color: #333; margin-top: 20px;">
    <li>In addition to SES, it can also model the 'Trend' component. However, it should not have seasonality.</li>
    <li>With this method, we can model both the 'Level' and 'Trend' components.</li>
</ul>

<h3 style="color: #3498db; font-size: 1.8em; margin-top: 30px;">Triple Exponential Smoothing (TES - Holt-Winters)</h3>
<ul style="font-size: 1.2em; color: #333; margin-top: 20px;">
    <li>This is the most advanced of the exponential smoothing methods.</li>
    <li>It can model the 'Level', 'Trend', and 'Seasonality' components.</li>
    <li>It is suitable for both stationary series and series that contain trends and seasonality.</li>
</ul>

# Stationarity Test (Dickey-Fuller Test)

<div style="background-color: #f8f9fa; padding: 40px; border-radius: 20px; box-shadow: 0 0 15px rgba(0,0,0,0.1); font-family: 'Arial', sans-serif; line-height: 1.6; border-left: 5px solid #2c3e50;">

<h3 style="color: #2c3e50; font-size: 1.5em; margin-bottom: 20px;">üìä Stationarity Test (Dickey-Fuller Test)</h3>

<p style="font-size: 1.2em; color: #333; text-align: justify;">
In time series analysis, this is a statistical test used to determine if a dataset is stationary. Stationarity in a time series means that its mean and variance remain constant over time. A non-stationary time series should be made stationary before analysis since many time series models are designed for stationary data.
</p>

<p style="font-size: 1.2em; color: #333; text-align: justify;">
The <b>Dickey-Fuller Test</b> is quite popular in this context. It's a unit root test that checks the stationarity of a time series.
</p>

<p style="font-size: 1.2em; color: #333; text-align: justify;">
The basic logic behind the Dickey-Fuller test is:
<ul>
    <li>The null hypothesis (H‚ÇÄ) states that the time series is not stationary (has a unit root).</li>
    <li>The alternative hypothesis (H‚ÇÅ) states that the time series is stationary.</li>
</ul>
</p>

<p style="font-size: 1.2em; color: #333; text-align: justify;">
Based on the p-value obtained from the test:
<ul>
    <li>If the p-value is less than a chosen significance level (often 0.05 or 5%), then the null hypothesis is rejected, and it's concluded that the time series is stationary.</li>
    <li>If the p-value is greater than the chosen significance level, the null hypothesis cannot be rejected, leading to the conclusion that the time series is not stationary.</li>
</ul>
</p>

<p style="font-size: 1.2em; color: #333; text-align: justify;">
There are various methods to make a non-stationary time series stationary, for instance, differencing is one such method.
</p>

</div>

```python
def is_stationary(y):

    # "HO: Non-stationary"
    # "H1: Stationary"

    p_value = sm.tsa.stattools.adfuller(y)[1]
    if p_value < 0.05: 
        print(F"Result: Stationary (H0: non-stationary, p-value: {round(p_value, 3)})")
    else: 
        print(F"Result: Non-Stationary (H0: non-stationary, p-value: {round(p_value, 3)})")
```

```python
is_stationary(y)
```

<div style="background-color: #f5f7fa; padding: 30px; border-radius: 20px; border: 1px solid #d1d8e0; font-family: 'Arial', sans-serif; line-height: 1.6;">
    <p style="font-size: 1.2em; color: #2c3e50; text-align: justify;">
        The function below conducts a stationarity test on a time series and simultaneously provides detailed information by visualizing its components.
    </p>
</div>

```python
def ts_decompose(y, model="additive", stationary=False):
    result = seasonal_decompose(y, model=model)
    fig, axes = plt.subplots(4, 1, sharex=True, sharey=False)
    fig.set_figheight(10)
    fig.set_figwidth(15)

    axes[0].set_title("Decomposition for " + model + " model")
    axes[0].plot(y, 'k', label='Original ' + model)
    axes[0].legend(loc='upper left')

    axes[1].plot(result.trend, label='Trend')
    axes[1].legend(loc='upper left')

    axes[2].plot(result.seasonal, 'g', label='Seasonality & Mean: ' + str(round(result.seasonal.mean(), 4)))
    axes[2].legend(loc='upper left')

    axes[3].plot(result.resid, 'r', label='Residuals & Mean: ' + str(round(result.resid.mean(), 4)))
    axes[3].legend(loc='upper left')
    plt.show(block=True)

    if stationary:
        is_stationary(y)
```

    <h2 style="color: #2c3e50; border-bottom: 2px solid #3498db;">Single Exponential Smoothing (SES)</h2>
    <p style="font-size: 1.1em; color: #333; text-align: justify;">
        Single Exponential Smoothing (SES) is a method used for time series analysis and forecasting. It is particularly ideal for time series that do not exhibit a pronounced trend or seasonality. The primary purpose of this method is to reduce random fluctuations in the data and highlight the main trend or pattern. SES is primarily successful in modeling the 'Level' component, which can be thought of as the average of the series. Hence, SES is often expressed as "SES = Level".
    </p>
    <p style="font-size: 1.1em; color: #333; text-align: justify;">
        The main principle of SES is to create future estimates by taking the weighted average of past observations. These weights decline in a manner that assigns more weight to more recent observations and less weight to older ones.
    </p>
    <p style="font-size: 1.1em; color: #333; text-align: justify;">
        Mathematically, the estimate of SES is expressed as:
        \[
        \hat{y}_{t+1} = \alpha y_t + (1-\alpha) \hat{y}_t
        \]
    </p>
    <p style="font-size: 1.1em; color: #333; text-align: justify;">
        Where:
        <ul>
            <li>\( \hat{y}_{t+1} \) is the forecast for the next step (at t+1).</li>
            <li>\( y_t \) is the actual observation at time t.</li>
            <li>\( \hat{y}_t \) is the forecast made for time t.</li>
            <li>\( \alpha \) is the smoothing factor, a value between 0 and 1.</li>
        </ul>
    </p>
    <p style="font-size: 1.1em; color: #333; text-align: justify;">
        The \( \alpha \) factor controls how quickly the model responds to changes in the data. A high \( \alpha \) value means the model responds more to recent observations, while a low \( \alpha \) value means it responds more slowly. However, SES gives good results only when there is no pronounced trend or seasonality in the series.
    </p>
</div>

```python
def ses_optimizer(train, alphas, step=48):

    best_alpha, best_mae = None, float("inf")

    for alpha in alphas:
        ses_model = SimpleExpSmoothing(train).fit(smoothing_level=alpha)
        y_pred = ses_model.forecast(step)
        mae = mean_absolute_error(test, y_pred)

        if mae < best_mae:
            best_alpha, best_mae = alpha, mae

        print("alpha:", round(alpha, 2), "mae:", round(mae, 4))
    print("best_alpha:", round(best_alpha, 2), "best_mae:", round(best_mae, 4))
    return best_alpha, best_mae
```

    <h3 style="color: #2c3e50;">Double Exponential Smoothing (DES)</h3>
    <p>Double Exponential Smoothing, commonly referred to as DES, is a forecasting method used for time series data. This technique is particularly apt for time series with evident trends. While Single Exponential Smoothing (SES) focuses solely on modeling the series' level, DES can model both the level and the trend.</p>
    <p>The fundamental premise behind DES is the application of two exponentially weighted averages to forecast a future value: one for the level and another for the trend.</p>
    <p>Mathematically, DES is expressed as:</p>
    <ul>
        <li>Level Equation: \( l_t = \alpha y_t + (1 - \alpha) (l_{t-1} + b_{t-1}) \)</li>
        <li>Trend Equation: \( b_t = \beta (l_t - l_{t-1}) + (1 - \beta) b_{t-1} \)</li>
        <li>Forecasting Equation (for time step t+1): \( \hat{y}_{t+1} = l_t + b_t \)</li>
    </ul>
    <p>Where:</p>
    <ul>
        <li>\( y_t \): Observation at time t</li>
        <li>\( l_t \): Level component at time t</li>
        <li>\( b_t \): Trend component at time t</li>
        <li>\( \alpha \): Level smoothing factor (between 0 and 1)</li>
        <li>\( \beta \): Trend smoothing factor (between 0 and 1)</li>
        <li>\( \hat{y}_{t+1} \): Forecast for time step t+1</li>
    </ul>
    <p>This method is highly effective for identifying trends in a time series. However, to model other components such as seasonality, more advanced techniques like Triple Exponential Smoothing or Holt-Winters are necessary.</p>
</div>

```python
def des_optimizer(train, alphas, betas, step=48):
    best_alpha, best_beta, best_mae = None, None, float("inf")
    for alpha in alphas:
        for beta in betas:
            des_model = ExponentialSmoothing(train, trend="add").fit(smoothing_level=alpha, smoothing_slope=beta)
            y_pred = des_model.forecast(step)
            mae = mean_absolute_error(test, y_pred)
            if mae < best_mae:
                best_alpha, best_beta, best_mae = alpha, beta, mae
            print("alpha:", round(alpha, 2), "beta:", round(beta, 2), "mae:", round(mae, 4))
    print("best_alpha:", round(best_alpha, 2), "best_beta:", round(best_beta, 2), "best_mae:", round(best_mae, 4))
    return best_alpha, best_beta, best_mae
```

<div style="padding: 30px; background-color: #f4f6f9; border: 2px solid #d5d9dc; border-radius: 10px; font-family: Arial, sans-serif;">
    <h2 style="color: #2c3e50;">Triple Exponential Smoothing (Holt-Winters)</h2>
    <p style="text-align: justify;">
        Triple Exponential Smoothing, also known as the Holt-Winters method, is an advanced technique used for time series analysis and forecasting. This method is designed to model three components of a time series: level, trend, and seasonality.
    </p>
    <p style="text-align: justify;">
        The main idea behind this method is to construct future predictions using three separate exponentially weighted averages, corresponding to the level, trend, and seasonality components.
    </p>
    <p><strong>Mathematical Representation:</strong></p>
    <ol>
        <li><b>Level Equation</b>: \( l_t = \alpha \times (y_t - s_{t-p}) + (1 - \alpha) \times (l_{t-1} + b_{t-1}) \)</li>
        <li><b>Trend Equation</b>: \( b_t = \beta \times (l_t - l_{t-1}) + (1 - \beta) \times b_{t-1} \)</li>
        <li><b>Seasonality Equation</b>: \( s_t = \gamma \times (y_t - l_t) + (1 - \gamma) \times s_{t-p} \)</li>
        <li><b>Forecast Equation</b>: \( \hat{y}_{t+m} = l_t + m \times b_t + s_{t-p+1+(m-1) \mod p} \)</li>
    </ol>
    <p style="text-align: justify;">
        Where:<br>
        \( y_t \): Observation at time \( t \)<br>
        \( l_t \): Level component at time \( t \)<br>
        \( b_t \): Trend component at time \( t \)<br>
        \( s_t \): Seasonality component at time \( t \)<br>
        \( \alpha, \beta, \gamma \): Smoothing parameters for level, trend, and seasonality respectively<br>
        \( p \): Seasonality period (e.g., for a monthly time series, \( p = 12 \))<br>
        \( \hat{y}_{t+m} \): Forecast for \( m \) time steps into the future
    </p>
    <p style="text-align: justify;">
        The Holt-Winters method is particularly suitable for time series that exhibit both a trend and seasonality. Thus, it is frequently used for time series where both of these components are present.
    </p>
</div>

# Model Creating

```python
tes_model = ExponentialSmoothing(train,
                                 trend="add",
                                 seasonal="add",
                                 seasonal_periods=12).fit(smoothing_level=0.5,
                                                          smoothing_slope=0.5,
                                                          smoothing_seasonal=0.5)
```

```python
y_pred = tes_model.forecast(48)
```

# Hyperparameter Optimization

```python
alphas = betas = gammas = np.arange(0.20, 1, 0.10)
```

```python
abg = list(itertools.product(alphas, betas, gammas))
```

```python
def tes_optimizer(train, abg, step=48):
    best_alpha, best_beta, best_gamma, best_mae = None, None, None, float("inf")
    for comb in abg:
        tes_model = ExponentialSmoothing(train, trend="add", seasonal="add", seasonal_periods=12).\
            fit(smoothing_level=comb[0], smoothing_slope=comb[1], smoothing_seasonal=comb[2])
        y_pred = tes_model.forecast(step)
        mae = mean_absolute_error(test, y_pred)
        if mae < best_mae:
            best_alpha, best_beta, best_gamma, best_mae = comb[0], comb[1], comb[2], mae
        print([round(comb[0], 2), round(comb[1], 2), round(comb[2], 2), round(mae, 2)])

    print("best_alpha:", round(best_alpha, 2), "best_beta:", round(best_beta, 2), "best_gamma:", round(best_gamma, 2),
          "best_mae:", round(best_mae, 4))

    return best_alpha, best_beta, best_gamma, best_mae
```

```python
best_alpha, best_beta, best_gamma, best_mae = tes_optimizer(train, abg)
```

# Final TES Model

```python
final_tes_model = ExponentialSmoothing(train, trend="add", seasonal="add", seasonal_periods=12).\
            fit(smoothing_level=best_alpha, smoothing_trend=best_beta, smoothing_seasonal=best_gamma)
```

```python
y_pred = final_tes_model.forecast(48)
```
